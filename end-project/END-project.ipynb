### Description of the Data

In the experiment ([[1]](#Citations)), two male rhesus monkeys (referred to by `s` and `m`) were trained to view an array of images (stimuli) and locate one of many possible target images in an array of distractor images.

Both monkeys had a recording chamber implanted over the left hemisphere, and the recording of the neural responses was performed in an area of the brain called the inferior temporal cortex (IT) with a single electrode advanced into the brain through a guiding tube using a micropositioner. The images were displayed on a graphics workstation at a resolution of $1024 \times 768$, and an EyeLink II video tracking system was used to record the eye movements.

The part of the data set explored here comes from the first stage of the experiment where the isolated images were shown individually without any distractors and the neurons' responses to those images were collected. For each trial, the relevant part of data set contains the spike times in milliseconds (`spk_times`), the stimulus names (`stim_names`), the subject code of the monkey (`subj`), and the time stamps in milliseconds when the images appeared on the screen (`stimon`).

### Programming Tricks

##### The `apply` method of pandas DataFrame

This [method](http://pandas.pydata.org/pandas-docs/dev/generated/pandas.DataFrame.apply.html) allows us to apply a function to one or several columns of a pandas data frame to produce new values — in particular, a new column.

     df['rates'] = df[['spk_times', 'stimon']].apply(lambda x: get_rate(x[0], x[1] + start, x[1] + stop), axis=1)
     
In this line, we are taking two columns from the data frame and then, using the `apply` method, apply a function (here `get_rate`) to the individual values from the columns by row. The `lambda` expression takes a parameter `x`, which is a tuple containing a value from one row of the first column and another value from the same row of the second column. This allows us to combine the columns using custom functions similarly to how we can combine the columns using built-in Python operators. This trick was useful in creating a column of the response rates, which was then used in the creation of the figures.

##### Using `matplotlib.gridspec.GridSpec` to set the width ratios of subplots in a figure

The [`gridspec.GridSpec` function](http://matplotlib.org/api/gridspec_api.html) allows us to customize the widths of the individual subplots in a Matplotlib plot:

    gs = mpl.gridspec.GridSpec(1, 2, width_ratios=[1, 2]) 

In this line, we tell Matplotlib that we want one row and two columns, i.e. two subplots in a row — but not just any subplots! The `width_ratios` parameter makes the second subplot twice as wide as the first subplot. This allowed us to make the histograms of the averaged images wider than they would have been otherwise. The default settings would have made interpreting the histograms much more difficult since the peaks would have been too close to each other.

### Problems

The images included in the data set were provided in the PNG format and included a channel for alpha transparency. Failing to take into account the transparency channel resulted in the loss of color information when averaging the images. For some reason, most of the image area appeared in various shades of grey. Resolving this problem by dropping the transparency channel allowed us to properly average the stimuli without loss of color information.
